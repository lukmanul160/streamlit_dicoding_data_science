### Assessing Data

### Cleaning Data

Lakukan perubahan terkait jenis data yang ada dan lakukan clean drop feature/field yang tidak dibutuhkan


```python
# pada 2 dataset terdapat dtime dimana masih bernilai/jenis object
# df_hours
df_hours['dteday'] = pd.to_datetime(df_hours['dteday'])
```

```python
# lakukan perubahan terkait dengan nama field/feature yang ada dimana hal tersebut kurang informatif
# df_hours
df_hours = df_hours.rename(columns={'weathersit':'weather',
                       'yr':'year',
                       'mnth':'month',
                       'hr':'hour',
                       'hum':'humidity',
                       'cnt':'count'})
```

```python
# ubah data dimana data kategorikal
cols = ['season' , 'month' , 'hour' , 'holiday' , 'weekday' , 'workingday' , 'weather']
for col in cols:
    df_hours[col] = df_hours[col].astype('category')
```

## Exploratory Data Analysis (EDA)

Lakukan analisa data yang ada pada dataset yang tersedia dimana dilakukan pengecekan dari value dari setiap feature/field yang ada

```python
# Analisis dekripsi data perjam
df_hours.describe()
```

```python
df_hours.hist()
```

1. diketahui bahwa casual,registered memiliki right skewness
2. untuk count memiliki left skewnes

```python
corr = df_hours.corr()
plt.figure(figsize=(15,7))
sns.heatmap(corr, annot=True, annot_kws={'size':10})
```

Melihat kolerasi dimaan dapat dilihat apabila semakin mendekati 0/- maka kolerasi antar feature tidak berkolerasi begitu juga sebalikny

## Visualization & Explanatory Analysis

```python
# Plotting Seaborn 
sns.pairplot(df_hours, diag_kind='kde')
plt.show()
```

berdasarkan plotting diatas dapat diketahui bahwa registered,casual memiliki hubungan regresi terhadap count dan terdapat hubungan regresi antara temp dan atemp


### Pertanyaan 1: Kapan waktu(jam) yang sering terjadi peminjaman/rent bike sharing yang ada

```python
# plotting berdasarkan waktu(hour)
fig, ax = plt.subplots(figsize=(20,5))
sns.pointplot(data=df_hours, x='hour', y='count', hue='weekday')
ax.set(title='Plotting berdasarkan banyak peminjaman/rent berdsarakan waktu dan hari')
```

### Pertanyaan 2: Berapa banyak dalam satu hari dan bulan dilakukan peminjaman rent bike sharing

```python
fig, ax = plt.subplots(figsize=(20,5))
sns.barplot(data=df_hours, x='weekday', y='count', ax=ax)
ax.set(title='total peminjaman sepeda berdasarkan hari kerja')
```

```python
fig, ax = plt.subplots(figsize=(20,5))
sns.barplot(data=df_hours, x='month', y='count', ax=ax)
ax.set(title='total peminjaman sepede berdasakran bulan')
```

### Pertanyaan 3 :pengaruh cuaca terkait dengan banyaknya peminjam/rent yang terjadi

```python
fig, ax = plt.subplots(figsize=(20,5))
sns.pointplot(data=df_hours, x='hour', y='count', hue='weather', ax=ax)
ax.set(title='Total Peminjaman sepeda berdasarkan Cuaca')
```

```python
fig, (ax1,ax2) = plt.subplots(ncols=2, figsize=(20,5))
sns.regplot(x=df_hours['temp'], y=df_hours['count'], ax=ax1 ,color='red')
ax1.set(title="hubungan antara suhu dan pengguna")
sns.regplot(x=df_hours['humidity'], y=df_hours['count'], ax=ax2)
ax2.set(title="hubungan antara kembebapan dengan pengguna")
```

- conclution pertanyaan 3 : berdasakran data yang ada apabila terjadi kenaikan temperaturn maka akan terjadi penuruan terkait dengan peminjaman/rent sepeda berdasarkan data yang ada apabila terjadi peningkatan kelembapan/humidity maka akan terjadi peningkatan terkait peminjaman/rent
